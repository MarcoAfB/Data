```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```python
data = np.random.exponential(6,  1000)
plt.hist(data, bins=30)
plt.show()
```

```python
'''É uma forma de colocar os dados dentro de uma escala sendo nesse caso [0,1]
é dito isso ajudar algoritmos que depende de distância entre parâmetros pois
agora não haverá desproporção muito grande entre as observações
Será feito o teste alguns modelos em diferentes data-set'''
def min_max(data_set):
    minn = min(data_set)
    maxx = max(data_set)
    new_data = []

    for i in data_set:
        new_data.append((i-minn)/(maxx-minn))

    return new_data
```

```python
minmax_data = min_max(data)
plt.hist(minmax_data, bins=30)
plt.show()
```

```python
'''Mesma ideia que minmax porem aqui não precisamos ter a mínima e a máxima da variável
(coisa que em alguns casos não é possível obter), também sendo mais robusto quanto
a outliers'''
def zscore_normalization(data_set):
    mean = np.mean(data_set)
    std = np.std(data_set)
    new_data = []
    for i in data_set:
        new_data.append((i-mean)/std)

    return new_data
```

```python
zscore_data = min_max(data)
plt.hist(zscore_data, bins=30)
plt.show()
```

```python
'''Diferente das anteriores box_cox busca alterar a distribuição da 
variável para gaussian também é um processo que pode melhorar o poder
preditivo do modelo ainda mais nos casos que eles tem como suposição 
os dados terem a distribuição gaussian'''
def box_cox(data, w):
    new_data = []
    for i in data:
        new_data.append((i**(w-1))/w)

    return new_data
```

```python
box = box_cox(data, 1.27)
plt.hist(box, bins=30)
plt.show()
```
